Feature Scaling

```python
from scipy.stats import probplot
```

```python
#Se os dados estiverem em distribuicao normal, eh melhor padroniza-los
#Se a distribuiÃ§Ã£o nao for Gaussiana (normal) ou o desvio padrao for muito pequeno, eh melhor normalizar
fig = plt.figure(figsize=[10, 10])
fig.subplots_adjust(hspace=0.45, wspace=0.45)
scaling_columns = ['PhysHlth', 'MentHlth', 'BMI']
x = 1

for column in scaling_columns:
  ax = fig.add_subplot(2, 2, x)
  probplot(df[column], fit=True, plot=ax)
  plt.title(f'Distribution of {column}')
  x = x + 1

plt.show()
```

```python
#Nas tres colunas, a distribuicao nao eh gaussiana; por isso, eh feita a normalizacao
from sklearn.preprocessing import MinMaxScaler

for column in scaling_columns:
  scaler = MinMaxScaler()
  df[column] = scaler.fit_transform(df[[column]])
```

Chi-squared Test

```python
from scipy.stats import chi2_contingency
```

```python
def chi_squared_test(data, column1, column2, alpha=0.05):

    contingency_table = pd.crosstab(data[column1], data[column2])

    test_results = chi2_contingency(contingency_table)
    p_value = test_results[1]

    if p_value <= alpha:
      H0_rejected = True
    else:
      H0_rejected = False

    return H0_rejected
```

```python
#O Teste do Qui-quadrado consegue comparar apenas variaveis categoricas
#Age, Education e Income sao dados ordinais; por isso, nao serao colocados nesse teste
categorical_columns = ['HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke',
                       'HeartDiseaseorAttack', 'PhysActivity', 'Fruits',
                       'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare',
                       'NoDocbcCost', 'GenHlth', 'DiffWalk', 'Sex']

chi_squared_test_results = {}

for column in categorical_columns:
  chi_squared_test_results[column] = chi_squared_test(df, 'Diabetes_binary', column)

chi_squared_test_results = pd.DataFrame(chi_squared_test_results, index=['H0 Rejected']).T
display(chi_squared_test_results)
```

Drop The Least Relevant Columns

```python
#Eliminacoes baseadas na correlacao e no Teste do Qui-quadrado
df = df.drop(columns=['Fruits', 'NoDocbcCost', 'AnyHealthcare'])
```

Balancing And Splitting The Dataset

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Diabetes_binary'])
```

```python
Y = df['Diabetes_binary']
```

```python
over_sampler = RandomOverSampler(sampling_strategy='minority', random_state=42)
X_balanced, Y_balanced = over_sampler.fit_resample(X,Y)
```

```python
plt.figure(figsize=(10,6))
plt.pie(Y_balanced.value_counts(), labels=['No diabetes', 'Diabetes'], autopct='%1.2f%%')
plt.title('Balanced Dataset: Diabetes Percentage')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_balanced, Y_balanced, test_size=0.2, random_state=42)
```

XGBoost Training

```python
from xgboost import XGBClassifier
```

```python
XGBoostModel = XGBClassifier(n_estimators=500, objective='binary:logistic', random_state=42)
```

```python
XGBoostModel.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
XGBoostPredictions = XGBoostModel.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, XGBoostPredictions)
report = classification_report(Y_test, XGBoostPredictions)
confusion = confusion_matrix(Y_test, XGBoostPredictions)

print("Accuracy:\n", accuracy)
print("Confusion Matrix:\n", confusion)
print("Relatory:\n", report)
```
